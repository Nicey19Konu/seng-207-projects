# Cleaning Dataset
###  We would clean the dataset by removing null values and duplicates

### Finding out if there are null values in the entire dataset


#### To find out if there are null values in the entire dataset, we will find the general info of the data if there are discrepancies
#### in the number of non null values, then we can conclude that there are null values that needs to be removed

```python
dataset.info()
```

### It can be seen from the info that there are discrepancies in the non-null counts, hence this confirms that there are null values in the dataset

### Filling the null values using the means

```python
meanfd=dataset['fixed acidity'].mean()
```

```python
meanfd=dataset['fixed acidity'].mean()
```

```python
meanfd
```

```python
meanvd=dataset['volatile acidity'].mean()
```

```python
meancd=dataset['citric acid'].mean()
meanrs=dataset['residual sugar'].mean()
meanc=dataset['chlorides'].mean()
meanfsd=dataset['free sulfur dioxide'].mean()
meantsd=dataset['total sulfur dioxide'].mean()
meand=dataset['density'].mean()
meanpH=dataset['pH'].mean()
means=dataset['sulphates'].mean()
meana=dataset['alcohol'].mean()
meanq=dataset['quality'].mean()
```

```python
dataset['quality']=dataset['quality'].fillna(meanq)
```

```python
dataset['fixed acidity']=dataset['fixed acidity'].fillna(meanfd)
dataset['volatile acidity']=dataset['volatile acidity'].fillna(meanvd)
dataset['citric acid']=dataset['citric acid'].fillna(meancd)
dataset['residual sugar']=dataset['residual sugar'].fillna(meanrs)
dataset['chlorides']=dataset['chlorides'].fillna(meanc)
dataset['free sulfur dioxide']=dataset['free sulfur dioxide'].fillna(meanfsd)
dataset['total sulfur dioxide']=dataset['total sulfur dioxide'].fillna(meantsd)
dataset['density']=dataset['density'].fillna(meand)
dataset['pH']=dataset['pH'].fillna(meanpH)
dataset['sulphates']=dataset['sulphates'].fillna(means)
dataset['alcohol']=dataset['alcohol'].fillna(meana)
```

```python
dataset
```

```python
dataset.info()
```

#### The method used was the fillna()approach using the mean. In this method, all the várious columns with various null values were analysed in detail and the mean values for the various values in each column was calculated. The mean for the various columns were then used to fill the respective null values for each of the columns'''

### Pie Chart
##### Here, I use pie chart to  find the percentage relations between the means of the various columns

```python
MeanValues=[meanfd,meanvd,meancd,meanrs,meanc,meanfsd,meantsd,meand,meanpH,means,meana,meanq]
MeanLabels=['FixedAcidity','VolatileAcidity','CitricACid','ResidualSugar','Chlorides','FreeSulfurDioxide','TotalSulfurDioxide',
            'Density','pH','Sulphates','alcohol','quality']
plt.pie(MeanValues,startangle=90,autopct='%1.2f',labels= MeanLabels, explode=(0,2,0,4,1,0,1,2,3,0,0,0))

plt.show()
```

#### HIDDEN INFERENCE FROM THE PIE CHART
##### The exploded pie chart above explains the percentage of the means of the various columns in the entire dataset. It can be observed from the pie chart that in the comparison of the columns of the dataset, the mean of total Sulfur Dioxide occupies the largest portion of the chart with the chlorides occupying the least portion of the chart
